
# Jupyter Notebook - Configurações


# Alterar pasta inicial:

No prompt de comando:

jupyter notebook --generate-config
* É criado o arquivo jupyter_notebook_config.py na pasta Users: "C:\Varnei\.jupyter"

Editar esse arquivo criado:
The directory to use for notebooks and kernels.
Default:
c.NotebookApp.notebook_dir = ''
Alterar para a pasta desejada:
c.NotebookApp.notebook_dir = 'D:\_ODFB0_VAM\OneDrive - vamti\_GitHub'

Depois, alterar as configurações do atalho do "Jupyter Notebook (Anaconda3)":
No campo "Destino", alterar, no final da linha, para a pasta desejada:
C:\ProgramData\Anaconda3\python.exe C:\ProgramData\Anaconda3\cwp.py C:\ProgramData\Anaconda3 C:\ProgramData\Anaconda3\python.exe C:\ProgramData\Anaconda3\Scripts\jupyter-notebook-script.py "D:\_ODFB0_VAM\OneDrive - vamti\_GitHub"
